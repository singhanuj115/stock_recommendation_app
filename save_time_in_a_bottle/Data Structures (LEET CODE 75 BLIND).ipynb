```python

# Very important!
def find_median_sorted_array(nums1, nums2):
    a, b = nums1, nums2

    total = len(nums1) + len(nums2)
    half = total//2
    
    if len(b)<len(a):
        a,b = b,a
    
    # left and right pointer for A
    l,r = 0,len(a) -1
    
    while True:
        median_a = (l + r)//2  # A
        median_b = half-median_a-2 #B
        
        a_left = a[median_a] if median_a>=0 else float("-infinity")
        a_right = a[median_a+1] if median_a+1 < len(a) else float("infinity")
        
        b_left = b[median_b] if median_b>=0 else float("-infinity")
        b_right = b[median_b+1] if median_b+1 < len(b) else float("infinity")
        
        # Correct partition
        if a_left <= b_right and b_left <= a_right:
            # odd
            if total % 2:
                return min(a_right, b_right)
            # even
            return (max(a_left, b_left)+min(a_right, b_right)) / 2
        
        elif a_left>b_right:
            r = median_a - 1
        elif b_left>a_right:
            l = median_a + 1
                
        
    
```

```python
# Minimum Window Substring
s = "ADOBECODEBANC"
t = "ABC"
if t == "": return ""

countT, window = {}, {}
for c in t:
    countT[c] = 1 + countT.get(c, 0)

have, need = 0, len(countT)
res, resLen = [-1, -1], float("infinity")
l = 0
for r in range(len(s)):
    c = s[r]
    window[c] = 1 + window.get(c, 0)

    if c in countT and window[c] == countT[c]:
        have += 1

    while have == need:
        if (r - l + 1) < resLen:
            res = [l, r]
            resLen = (r - l + 1)
            # pop from left window
        window[s[l]] -= 1
        if s[l] in countT and window[s[l]] < countT[s[l]]:
            have -= 1
        l += 1

l, r = res
return s[l:r + 1] if resLen != float("infinity") else ""
```

```python
# valid anagram
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        count_s_map, count_t_map = {}, {}
        
        for i in range(len(s)):
            count_s_map[s[i]] = 1+count_s_map.get(s[i], 0)
            
            count_t_map[t[i]] = 1+count_t_map.get(t[i], 0)
        
        for char in count_s_map:
            if count_s_map[char] != count_t_map.get(char, 0):
                return False
        
        return True
```

```python
# group anagram
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        res = defaultdict(list)
        
        for s in strs:
            
            count = [0]*26
            
            for c in s:
                count[ord(c)-ord("a")] +=1
            
            res[tuple(count)].append(s)
        
        return res.values()
```

```python
# valid paranthesis
class Solution:
    def isValid(self, s: str) -> bool:
        close_to_open = {")":"(",
                         "}": "{",
                        "]": "["}
        
        stack = []
        
        for c in s:
            # check open or close paranthesis:
            if c in close_to_open:
                if stack and stack[-1]==close_to_open[c]:
                    stack.pop()
                else:
                    return False
                
            else:
                stack.append(c)
        
        return True if not stack else False
                
                
```

```python
# Palindrome
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l, r = 0, len(s)-1
        
        while l<r:
            while l<r and not self.alpha_num(s[l]):
                l+=1
            while r>l and not self.alpha_num(s[r]):
                r-=1
            
            if s[l].lower() != s[r].lower():
                return False
            
            l, r = l+1, r-1
        return True
        
    
    def alpha_num(self, c):
        return ((ord("A") <= ord(c) <= ord("Z")) or
                (ord("a") <= ord(c) <= ord("z")) or
                (ord("0") <= ord(c) <= ord("9")) 
               )
```

```python
from collections import deque


class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    """
    n -> pre
    n -> in
    n -> post
    
              3
            /   \
           9      20
                 /   \
               15     7
    """

    def pre_traversal(self, temp):
        if not temp:
            return

        print(temp.val)
        self.pre_traversal(temp.left)
        self.pre_traversal(temp.right)

    def in_traversal(self, temp):
        if not temp:
            return

        self.in_traversal(temp.left)
        print(temp.val)
        self.in_traversal(temp.right)

    def post_traversal(self, temp):
        if not temp:
            return

        self.post_traversal(temp.left)
        self.post_traversal(temp.right)
        print(temp.val)

    def recursive_dfs_max_depth(self, temp):
        if not temp:
            return 0

        return 1 + max(self.recursive_dfs_max_depth(temp.left),
                       self.recursive_dfs_max_depth(temp.right)
                       )

    def bfs_max_depth(self, temp):
        if not temp:
            return 0

        level = 0
        q = deque([temp])

        while q:
            for i in range(len(q)):
                node = q.popleft()
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            level += 1

        return level

    def max_sum_path(self, temp):
        res = [temp.val]

        # return max path sum without split
        def dfs(temp):
            if not temp:
                return 0

            left_max = dfs(temp.left)
            right_max = dfs(temp.right)
            left_max = max(0, left_max)
            right_max = max(0, right_max)

            # Compute max path with split
            res[0] = max(res[0], temp.val + left_max + right_max)

            return temp.val + max(left_max, right_max)

        dfs(temp)
        return res[0]

    def serialize(self, root):
        res = []

        def dfs(root):
            if not root:
                res.append("N")
                return

            res.append(str(root.val))
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return ",".join(res)

    def de_serialize(self, serialized_str):
        vals = serialized_str.split(",")
        self.i = 0

        def dfs():
            if vals[self.i] == "N":
                self.i += 1
                return None

            node = TreeNode(vals[self.i])
            self.i += 1
            node.left = dfs()
            node.right = dfs()
            return node

        return dfs()

    def validate_bst(self, root):
        if not root:
            return

        root_val = root.val
        if root_val < root.left.val or root_val > root.left.right:
            return False
        self.validate_bst(root.left)
        self.validate_bst(root.right)


# tree_obj = TreeNode(3, TreeNode(9), TreeNode(20, TreeNode(15), TreeNode(7)))
tree_obj = TreeNode(2, TreeNode(1), TreeNode(3))
# tree_obj.pre_traversal(tree_obj)
# print("")
# tree_obj.in_traversal(tree_obj)
# print("")
# tree_obj.post_traversal(tree_obj)
# max_depth = tree_obj.recursive_dfs_max_depth(tree_obj)
# print(max_depth)
# max_depth = tree_obj.bfs_max_depth(tree_obj)
# print(max_depth)
# max_path_sum = tree_obj.max_sum_path(tree_obj)
# print(max_path_sum)
serialized_str = tree_obj.serialize(tree_obj)
deserialize = tree_obj.de_serialize(serialized_str)
tree_obj.pre_traversal(deserialize)
```
