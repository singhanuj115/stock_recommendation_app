```python
import os
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


from collections import Counter
from sklearn import neighbors
from sklearn import svm
from sklearn.model_selection import train_test_split


from sklearn.ensemble import VotingClassifier, RandomForestClassifier
```

```python
abs_path = "/Users/anujsingh/Desktop/notebooks/Scrape data for NSE/stock-market-india/Data/"
```

```python
csv_path_list = os.listdir(abs_path)
```

```python
df = pd.read_csv(abs_path + csv_path_list[0])
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df = df[(df['timestamp'].dt.day == 12) & (df['timestamp'].dt.month == 5)  & (df['timestamp'].dt.hour == 10)]
```

```python
df['average_price'] = (df['high']+ df['low']+ df['close'])/3
df['average_price_volume'] = df['average_price']* df['volume']
df['average_price_volume'] = df['average_price_volume'].cumsum()
```

```python
df["cumulative_vol"] = df["volume"].cumsum()
```

```python
df['vwap'] = df["average_price_volume"]/df["cumulative_vol"]
```
